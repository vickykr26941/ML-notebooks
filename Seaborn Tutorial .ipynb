```python
sns.heatmap(df.corr())
# heatmap plot the correlation matrix 
# it will mostely used in feature selection 
```

### example with univariate Analysis

```python
# here bivariate analysis b/w tip and total_bill 
sns.jointplot(x='tip',y='total_bill',data = df,kind='hex')
```

```python
# as we see above mostely bill payed by people is aroun b/w 10 and 20 and mostely tip given b/w 2 and 4 $ or rupy
# the kind = hex means that the data points sowing in forms of hexagon 
```

```python
sns.jointplot(x = 'tip', y = 'total_bill',data = df,kind = 'scatter')
```

```python
sns.jointplot(x='tip',y = 'total_bill', data = df,kind = 'reg') # reg = regression 
```

```python
# it's drow regression line and probability density function 
```

### Pair plot 

###### A pair plot is also known as a scatterplot ,in which one variable in the same data row is matched with another variable's value,like this: Pairs are just elaborations on this showing all variables paired with all the other variables

```python
# we use when more then two independent features 
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='sex')
# hue parameter = we can classifie the scatter plot on the basis of catagorical variable
```

### Dist plot 

###### Dist plot helps us to check the distrubution of the columns features

```python
sns.distplot(df['tip'])
#histogram with probability density function 
```

```python
sns.distplot(df['tip'],kde=False)
# kde = kernal density estimation :if it is true is will show the pdf(probability density function also otherwise not)
```

```python
# we can specifie bins also 
sns.distplot(df['tip'],kde=False,bins = 10)
```

# Ploting for categorical features

###### boxplot
######  viollnplot

```python
sns.countplot(x=df['sex'],data=df)
```

```python
df['sex'].value_counts()
```

```python
sns.countplot(x=df['day'])
```

```python
df['day'].value_counts()
```

```python
sns.countplot(y='day',data = df)
```

```python
sns.barplot(x='sex',y='total_bill',data = df)
```

```python
sns.barplot(x='total_bill',y='smoker',data = df)
```

# box plot

```python
sns.boxplot(x='sex',y='total_bill',data = df)
```

```python
sns.boxplot(x='day',y='total_bill',data =df)
```

```python
sns.boxplot(data =df,orient = 'v') # v for vertical 
#here by default it is taken all the numerical features
```

```python
# categorize my data based on some other categories
sns.boxplot(x='total_bill',y='day',hue='smoker',data = df)
```

# violin plot

###### violin plot help us to see both the distrubution of the data in terms of kernel density estimation and box plot

```python
sns.violinplot(x='total_bill',y='day',data = df,palette = 'rainbow')
```

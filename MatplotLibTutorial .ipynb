```python
x = np.arange(0,10*np.pi,0.1)
y = np.sin(x)
plt.title('signe wave form')
plt.plot(x,y)
plt.show()
```

```python
x = np.arange(0,10*np.pi,0.1)
y = np.cos(x)
plt.title('cos wave form')
plt.plot(x,y)
plt.show()
```

```python
#linear equation 
# 3*x + 4*y
x = np.arange(10,200)
y = np.arange(20,210)
z = 3*x + 4*y
```

```python
plt.plot(x,y,'r',label = 'x,y')
plt.plot(x,z,'b',label = '3*x + 4*y')
plt.title('linear graph 3*x + 4*y')
plt.show()
```

```python
x = np.arange(0,5*np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.subplot(2,1,1)
plt.plot(x,y_sin,'r--')
plt.title('sin curve')

plt.subplot(2,1,2)
plt.plot(x,y_cos,'b--')
plt.title('cos curve')
plt.show()
```

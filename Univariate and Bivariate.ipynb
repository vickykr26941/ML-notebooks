
# Univariate, Bivariate and Multivariate 

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv("tableconvert_csv_4jcpx8.csv")
```

```python
df.head()
```

```python
# rename the columns names of the columns
df.columns = ['sepal_length','sepal_width','petal_length','petal_width','species']
```

```python
df.head()
```

```python
# find the total types of species
df['species'].value_counts()
```

```python
df.shape
```

### Univariate Analysis

```python
# we use only take single data series and try to plot and see the behaviour of 
# data that's univariate analysis
```

```python


# take all kind of flowers seperately seperately 
df_setosa = df.loc[df['species']=='Setosa']
df_virginica = df.loc[df['species']=='Virginica']
df_versicolor = df.loc[df['species']=='Versicolor']
```

```python
plt.plot(df['sepal_length'],np.zeros_like(df['sepal_length']),'o')
```

```python
# but if we want to see the each speciece in different colors to see the relation 
plt.plot(df_setosa['sepal_length'],np.zeros_like(df_setosa['sepal_length']),'o')
plt.plot(df_virginica['sepal_length'],np.zeros_like(df_virginica['sepal_length']),'o')
plt.plot(df_versicolor['sepal_length'],np.zeros_like(df_versicolor['sepal_length']),'o')
```

```python
# now as we can see the "setosa" have blue color and in green color virginica
# and in yellow color versicolor now we can see the anlyse easily 
```

```python
# let try to plot the same graph as above without spliting the data using 
# matplotlib and seaborn 
grouped = df.groupby('species')
for key,group in grouped:
    plt.plot(group['sepal_length'],np.zeros_like(group['sepal_length']),'o')
    
# using groupy by pandas series method
```

# Bivariate Analysis

```python
# in bivariate analysis we take two features and try to relate them and analyse
# the relation between them by various plots etc that's bivarite analysis
```

```python
# ploting using seaborn 
sns.FacetGrid(df,hue = 'species',size = 5).map(plt.scatter,"sepal_length","sepal_width").add_legend()
```

```python
# bivariate analysis using joint plot
sns.jointplot(x ='sepal_length',y = 'sepal_width',data = df,kind = 'scatter',hue = 'species')
```

### Multivariate Analysis

```python
# we take more then two fertures to analyse the relation between them
# that is basically multivariate analysis
```

```python
sns.pairplot(df,hue = 'species')
```

```python
# we have did univariate analysis by ploting the line to the parallel to 
# the x axis but if data points is large then that can not help us to figure
# out the relation between the data points so now we can use histogram for
# univariate analysis
```

```python
# histogram usig matplot lib
plt.hist(df['sepal_length'],bins = 30)
plt.show()
```

```python
# ploting histogram using distplot of seaboarn
sns.distplot(df['sepal_length'],kde= None,bins = 30)
```

```python
# plot histogram for each species
grouped = df.groupby('species')
for key ,group in grouped:
    print(key)
    if key=='Setosa':
        plt.hist(group['sepal_length'],color = 'blue',bins = 25)
```

```python
for key ,group in grouped:
    if key=='Versicolor':
        plt.hist(group['sepal_length'],color = 'blue',bins = 25)
```

```python
for key ,group in grouped:
    if key=='Virginica':
        plt.hist(group['sepal_length'],color = 'blue',bins = 25)
```
